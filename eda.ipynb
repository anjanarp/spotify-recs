```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

combined_songs_df = pd.read_csv('combined_songs_data.csv')
```

```python
combined_songs_df.info()
```

```python
numerical_features = [
    'popularity', 'danceability', 'energy', 'key', 
    'loudness', 'speechiness', 'acousticness', 
    'liveness', 'Valence', 'tempo'
]

# calculate the correlation matrix
correlation_matrix = combined_songs_df[numerical_features].corr()

# plot the correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

numerical_data = combined_songs_df[numerical_features]

vif_data = pd.DataFrame()
vif_data['Feature'] = numerical_data.columns
vif_data['VIF'] = [variance_inflation_factor(numerical_data.values, i) for i in range(len(numerical_data.columns))]

print(vif_data)
```
